```python
import numpy as np
import pandas as pd
```

```python
a=[]
for i in ["PS1","PS2","PS3","PS4","PS5","PS6",
         "TS1","TS2","TS3","TS4","FS1","FS2",
          "EPS1", "VS1", "SE"
         ]:
    sensors=np.genfromtxt(
        r"/Users/bagusemeraldo/Desktop/Pacmann/DS/Introduction to Machine Learning I/Project/hydraulic"
        +"/"+i+".txt")
    a.append(sensors)
ps1,ps2,ps3,ps4,ps5,ps6,ts1,ts2,ts3,ts4,fs1,fs2,eps1,vs1,se=a
```

```python
#calculate the mean of every sensors
ps1 = ps1.mean(axis=1)
ps2 = ps2.mean(axis=1)
ps3 = ps3.mean(axis=1)
ps4 = ps4.mean(axis=1)
ps5 = ps5.mean(axis=1)
ps6 = ps6.mean(axis=1)

ts1 = ts1.mean(axis=1)
ts2 = ts2.mean(axis=1)
ts3 = ts3.mean(axis=1)
ts4 = ts4.mean(axis=1)

fs1 = fs1.mean(axis=1)
fs2 = fs2.mean(axis=1)

eps1 = eps1.mean(axis=1)

vs1 = vs1.mean(axis=1)

se = se.mean(axis=1)
```

```python
df=pd.DataFrame({"PS1":ps1,"PS2":ps2,"PS3":ps3,"PS4":ps4,"PS5":ps5,"PS6":ps6,
                "TS1":ts1,"TS2":ts2,"TS3":ts3,"TS4":ts4,"FS1":fs1,"FS2":fs2,
                 "EPS1":eps1, "VS1":vs1, "SE":se})
```
